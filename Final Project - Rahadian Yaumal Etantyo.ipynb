## **Metric Evaluation**
| ACTUAL \ PREDICTED | Not Churn (0) | Churn (1) |
| ------------------ | ------------- | --------- |
| Not Churn (0)      | TN            | FP        |
| Churn (1)          | FN            | TP        |

Keterangan:<br>
- TN (True Negatif) = Aktualnya Not Churn dan machine learning memprediksi not churn
- FP (False Positif) = Aktualnya Not Churn dan machine learning memprediksi churn
- FN (False Negatif) = Aktualnya Churn dan machine learning memprediksi not churn
- TP (True Positif) = Aktualnya Churn dan machine learning memprediksi churn <br>

Type Eror:<br>
- Type Error 1 (FP) False Positif = Biaya marketing yang dikeluarkan membengkak dan sia sia (3 juta/orang).
- Type Error 2 (FN) False Negatif = Potensi kehilangan pendapatan (6 juta/orang)

<p align="justify"> 
Tujuan utama analisis perilaku pelanggan adalah mengidentifikasi secara akurat mana pelanggan yang berisiko churn dan mana yang tidak, sehingga perusahaan dapat mengoptimalkan strategi retensi dengan biaya yang efisien. Fokus dalam machine learning ini adalah meminimalkan nilai False Negatif, karena kegagalan mendeteksi pelanggan yang benar-benar churn akan menyebabkan hilangnya potensi pendapatan akibat kehilangan pelanggan serta biaya yang dikeluarkan untuk menarik pelanggan kembali lebih besar daripada biaya untuk mempertahankan pelanggan loyal. Dalam konteks ini, metric evaluasi utama yang digunakan adalah F2 Score. Pemakaian F2 Score dipilih karena dalam kasus prediksi churn, kesalahan false negative (pelanggan yang sebenarnya churn tetapi diprediksi tidak churn) lebih merugikan perusahaan dibanding false positive. F2 memberikan bobot lebih besar pada recall, sehingga model lebih fokus menangkap sebanyak mungkin pelanggan berisiko churn untuk meminimalkan potensi kehilangan biaya yang lebih besar baik pendapatan maupun biaya untuk menarik kembali pelanggan yang churn.
</p>




## **Rumus F2 Score**
Rumus umum Fscore: <br>
$$F_{\beta} = \frac{(1 + \beta^{2}) \cdot (Precision \cdot Recall)}{(\beta^{2} \cdot Precision) + Recall}
$$

Rumus F2 Score : <br>
$$
F_{2} = \frac{5 \cdot Precision \cdot Recall}{(4 \cdot Precision) + Recall}
$$

## **Define X & Y**

```python
X = df_new.drop(columns='Churn')
y = df_new['Churn']
```

Hal ini dilakukan untuk memisahkan kolom target (y) yang akan diprediksi  dengan feature(X) yang akan dipelajari oleh machine learning untuk menemukan pola yang ada dan memberikan prediksi yang akurat. 

## **Data Splitting**

```python
X_train, X_test, y_train, y_test = train_test_split(X, 
                 y,
                 test_size=0.2,
                 random_state=0,
                 stratify=y)
```

Data Splitting  ini bertujuan untuk membagi dataset menjadi data latih (80%) dan data uji (20%) dengan stratifikasi berdasarkan variabel target (churn). Dengan adanya stratify=y, jumlah perbandingan antara pelanggan churn dan tidak churn tetap sama di kedua bagian data, sehingga hasil uji lebih adil. Sementara itu, random_state dipakai agar pembagian data selalu sama setiap kali kode dijalankan.

## **Data Preprocessing**

```python
listitem = []
for col in df_new.columns:
    listitem.append([col,df_new[col].dtypes,df_new[col].isna().sum(),df_new[col].nunique(),df_new[col].unique()])

data_kolom = pd.DataFrame(data=listitem, columns=['Nama Kolom','Jenis Data','Jumlah kolom kosong','Jumlah Nilai Unik Tiap Kolom','Nilai Unik'])
data_kolom
```

```python
pipe_iter_robust = Pipeline([
    ('SimpleImput',SimpleImputer(strategy="median")),
    ('scaling',RobustScaler())
])
Transformer = ColumnTransformer([
    ('pipe iter robust',pipe_iter_robust,['Tenure','WarehouseToHome','HourSpendOnApp','OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder']),
    ('scaling',RobustScaler(),['CashbackAmount']),
    ('encode',OneHotEncoder(),['PreferredLoginDevice','PreferredPaymentMode','Gender','PreferedOrderCat','MaritalStatus'])
],remainder='passthrough') 
```

<p align="Justify">
Pada tahap data preprocessing dilakukan penginputan data missing value pada kolom'Tenure','WarehouseToHome','HourSpendOnApp' 'OrderAmountHikeFromlastYear','CouponUsed','OrderCount','DaySinceLastOrder' menggunakan strategi median karena lebih aman untuk menghindari outlier. Kemudian juga dilakukan scaling pada kolom tersebut menggunakan robust scaler dengan tujuan menstandarkan skala data numerik supaya semua fitur berada dalam rentang yang sebanding, tanpa terlalu dipengaruhi oleh outlier. Penerapan robust scaler juga dilakukan pada kolom cashbackamount karena banyak data outlier di kolom tersebut. Pada data jenis category dihandle dengan cara mengencoding data tersebut dengan bantuan onehotencoder. Encoding ini bertujuan agar data berubah menjadi angka dan machine learning mampu memproses data ini. Pemilihan onehotencoder karena data category tersebut nilai unik tidak terlalu banyak sehingga onehotencoder menjadi pilihan
</p>

## **CrossValidation**

### *Define Model*

```python
tree = DecisionTreeClassifier(random_state=0)
knn = KNeighborsClassifier()
logreg = LogisticRegression(random_state=0)
bagging = BaggingClassifier(random_state=0)
rf = RandomForestClassifier(random_state=0)
aboost = AdaBoostClassifier(random_state=0)
gboost = GradientBoostingClassifier(random_state=0)
xboost = XGBClassifier()
```

```python
model = [tree, knn, logreg, bagging, rf, aboost, gboost, xboost]
all_score = []
mean_score = []
std_score = []
model_names = []
f2_scorer = make_scorer(fbeta_score, beta=2)

for mod in model:
    pipe_prep_model = ImbPipeline([
        ('transformer', Transformer),
        ('balancing', SMOTE(random_state=0)),
        ('modeling', mod)
    ])
    
    crossval = cross_val_score(estimator=pipe_prep_model,
                               X=X_train,
                               y=y_train,
                               scoring=f2_scorer,
                               cv=5)
    all_score.append(crossval.round(2))
    mean_score.append(crossval.mean())
    std_score.append(crossval.std())
```

```python
table_score = pd.DataFrame({
    'Nama Model':model,
    'Score All' :all_score,
    'Rata - Rata Score': mean_score,
    'std_score':std_score
})
table_score
```

<p align="Justify">
Dari data x dan y yang dilatih dengan beberapa model machine learning, diperoleh model terbaik adalah XGBClassifier dengan rata - rata score 0.867520. XGBClassifier menjadi yang terbaik karena rata - rata score yang diperoleh lebih tinggi dibanding dengan model yang lain.
</p>

### *Cara Kerja Model yang Dipilih (XGBClassifier)*

<p align="justify">
XGBoost Classifier bekerja dengan pendekatan gradient boosting, yaitu membangun sejumlah pohon keputusan (decision tree) secara bertahap untuk memperbaiki kesalahan prediksi dari model sebelumnya. Proses dimulai dari prediksi awal yang sederhana, kemudian pada setiap iterasi XGBoost menghitung gradien dari loss function untuk mengetahui arah perbaikan yang diperlukan. Pohon keputusan baru kemudian dilatih untuk meminimalkan kesalahan tersebut dan hasilnya digabungkan dengan model sebelumnya menggunakan bobot tertentu melalui learning rate. Keunggulan XGBoost terletak pada adanya regularisasi (L1/L2) yang mampu mengendalikan kompleksitas model sehingga mengurangi risiko overfitting. Setelah seluruh iterasi selesai, gabungan pohon keputusan menghasilkan probabilitas prediksi yang kemudian dikonversi menjadi kelas akhir, misalnya pada kasus ini adalah churn atau tidak churn
</p>

## **Hyperparameter Tuning**

```python
pipe_prep_tune = ImbPipeline([
    ('transformer',Transformer),
    ('balancing',SMOTE(random_state=0)),
    ('model',xboost)
])
param_grid = {
    'model__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'model__n_estimators': [200, 500, 800],
    'model__max_depth': [3, 4, 5, 6, 8],
    'model__min_child_weight': [1, 3, 5],
    'model__gamma': [0, 0.1, 0.3, 0.5],
    'model__subsample': [0.7, 0.8, 0.9, 1.0],
    'model__colsample_bytree': [0.6, 0.8, 1.0],
    'model__reg_alpha': [0, 0.01, 0.1, 1],
    'model__reg_lambda': [1, 5, 10]
}

rand_search = RandomizedSearchCV(
    estimator= pipe_prep_tune,
    param_distributions = param_grid,
    cv = 5,
    n_jobs=-1,
    scoring = f2_scorer,
    n_iter=150
)
rand_search
```

```python
rand_search.fit(X_train, y_train)
```

```python
rand_search.best_estimator_
```

```python
rand_search.best_params_
```

Hasil dari hyperparameter tuning diperoleh beberapa parameter terbaik yang akan digunakan untuk membandingkan score machine learning apakah lebih bagus setelah dilakukan tuning atau sebelum tuning

```python
rand_search.best_score_
```

Merupakan score terbaik yang diperoleh machine learning ketika mencoba menggunakan semua kombinasi parameter yang ada
